```python
data_path = "../data"
group_list = ["group1A", "group2A"]
for group in group_list:
    
    group_path = os.path.join(data_path, group)
    group_dir = os.listdir(group_path)
    antor_path_list = []
    for name in group_dir:
        if "annotator" in name:
            antor_path_list.append(os.path.join(group_path, name))
#     print(antor_path_list)     
    Annotation_total = np.zeros((len(antor_path_list), len(os.listdir(antor_path_list[0]))-1))
    k = 0
    for antor_path in antor_path_list:
        anno_list = sorted(os.listdir(antor_path))[1:]
        Annotation_array = np.zeros(len(anno_list))
        for i in range(len(anno_list)):
            with open(os.path.join(antor_path,anno_list[i]),'rb') as f:
                aList = f.readlines()
            f.close()
            if b"Correct" in aList[10]:
                Annotation_array[i] = 1
#             if b"Incomplete" in aList[10]:
#                 Annotation_array[i] = np.nan
        print(antor_path, Annotation_array)
#         print(Annotation_total)
        Annotation_total[k] = Annotation_array
        k += 1
    save_path = "../results/pkls/"+group+"_annotation.npy"
    with open(save_path, 'wb') as g:
        np.save(g,Annotation_total)
    g.close()
    print(group_path, Annotation_total)
```

# Krippendorff Alpha and Cohen's Kappa

```python
import krippendorff
from sklearn import metrics
import numpy as np


g1_anno = np.load("../results/pkls/group1A_annotation.npy")
g1_test = g1_anno[1,:]
g1_test = np.vstack((g1_test,g1_anno[1,:]))
# print(g1_test)
g1_anno14 = g1_anno[(0,1),:]
g1_anno15 = g1_anno[(0,2),:]
g1_anno45 = g1_anno[(1,2),:]
# print(g1_14)
# print(g1_15)

g2_anno = np.load("../results/pkls/group2A_annotation.npy")
g2_anno14 = g2_anno[(0,1),:]
g2_anno15 = g2_anno[(0,2),:]
g2_anno45 = g2_anno[(1,2),:]

print("-------------------------G1A----------------------")
krippen_g1A = krippendorff.alpha(reliability_data = g1_anno, level_of_measurement="nominal")
krippen_g1A_14 = krippendorff.alpha(reliability_data = g1_anno14, level_of_measurement="nominal")
krippen_g1A_15 = krippendorff.alpha(reliability_data = g1_anno15, level_of_measurement="nominal")
krippen_g1A_45 = krippendorff.alpha(reliability_data = g1_anno45, level_of_measurement="nominal")

cohen_g1A_14 = metrics.cohen_kappa_score(g1_anno[0,:],g1_anno[1,:])
cohen_g1A_15 = metrics.cohen_kappa_score(g1_anno[0,:],g1_anno[2,:])
cohen_g1A_45 = metrics.cohen_kappa_score(g1_anno[1,:],g1_anno[2,:])
print("Krippendorff's alpha for three annotator: ", krippen_g1A)
print("Krippendorff's alpha for annotator1 and 4: ", krippen_g1A_14)
print("Cohen's kappa for annotator1 and 4: ", cohen_g1A_14)
print("Krippendorff's alpha for annotator1 and 5: ", krippen_g1A_15)
print("Cohen's kappa for annotator1 and 5: ", cohen_g1A_15)
print("Krippendorff's alpha for annotator4 and 5: ", krippen_g1A_45)
print("Cohen's kappa for annotator4 and 5: ", cohen_g1A_45)
print("-------------------------G2A----------------------")
krippen_g2A = krippendorff.alpha(reliability_data = g2_anno, level_of_measurement="nominal")
krippen_g2A_14 = krippendorff.alpha(reliability_data = g2_anno14, level_of_measurement="nominal")
krippen_g2A_15 = krippendorff.alpha(reliability_data = g2_anno15,level_of_measurement="nominal")
krippen_g2A_45 = krippendorff.alpha(reliability_data = g2_anno45, level_of_measurement="nominal")
cohen_g2A_14 = metrics.cohen_kappa_score(g2_anno[0,:],g2_anno[1,:])
cohen_g2A_15 = metrics.cohen_kappa_score(g2_anno[0,:],g2_anno[2,:])
cohen_g2A_45 = metrics.cohen_kappa_score(g2_anno[1,:],g2_anno[2,:])
print("Krippendorff's alpha for nominal metric: ", krippen_g2A)
print("Krippendorff's alpha for annotator1 and 4: ", krippen_g2A_14)
print("Cohen's kappa for annotator1 and 4: ", cohen_g2A_14)
print("Krippendorff's alpha for annotator1 and 5: ", krippen_g2A_15)
print("Cohen's kappa for annotator1 and 5: ", cohen_g2A_15)
print("Krippendorff's alpha for annotator4 and 5: ", krippen_g2A_45)
print("Cohen's kappa for annotator4 and 5: ", cohen_g2A_45)

# print("Krippendorff's alpha for interval metric: ", krippendorff.alpha(reliability_data=g2_anno))
```

# Write in csv

```python
import csv
csv_path = "../results/csvs/Agreements.csv"
field_order = ["Group", "Krippendorff Alpha", "Cohen Kappa"]
aType = ["group1A_total", "group1A_anno14","group1A_anno15","group1A_anno45",
        "group2A_total", "group2A_anno14","group2A_anno15","group2A_anno45"]
with open(csv_path,"w",encoding = "utf-8",newline = '') as f:
    writer = csv.DictWriter(f, field_order)
    writer.writeheader()
    writer.writerow(dict(zip(field_order, ["group1A_total", format(krippen_g1A,".4f"), np.nan])))
    writer.writerow(dict(zip(field_order, ["group1A_anno14", format(krippen_g1A_14,".4f"), format(cohen_g1A_14,".4f")])))
    writer.writerow(dict(zip(field_order, ["group1A_anno15", format(krippen_g1A_15,".4f"), format(cohen_g1A_15,".4f")])))
    writer.writerow(dict(zip(field_order, ["group1A_anno45", format(krippen_g1A_45,".4f"), format(cohen_g1A_45,".4f")])))
    
    writer.writerow(dict(zip(field_order, ["group2A_total", format(krippen_g2A,".4f"), np.nan])))
    writer.writerow(dict(zip(field_order, ["group2A_anno14", format(krippen_g2A_14,".4f"), format(cohen_g2A_14,".4f")])))
    writer.writerow(dict(zip(field_order, ["group2A_anno15", format(krippen_g2A_15,".4f"), format(cohen_g2A_15,".4f")])))
    writer.writerow(dict(zip(field_order, ["group2A_anno45", format(krippen_g2A_45,".4f"), format(cohen_g2A_45,".4f")])))
f.close()
```
